```python
import numpy as np

A = np.array([1, 2, 3])
B = np.array([4, 5, 6])
C = A+B
D = A*B
E = np.dot(A, B)
F = np.cross(A, B)

print(C)
print(D)
print(E)
print(F)

```

```python
import numpy as np

A = np.array([[4, 3],
              [1, 2]])

eigen = np.linalg.eig(A)

print(eigen)
```

```python
import numpy as np
from scipy import stats

q = [10,20,30,40,50]
mean = np.mean(q)
print("mean:",mean)

median = np.median(q)
print("median:",median)

mode = stats.mode(q)
print("mode:",mode)

var = np.var(q)
print("var:",var)

std = np.std(q)
print("std:",std)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

mu = 0
sigma = 1

# Generate data points
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 100)
y = stats.norm.pdf(x, mu, sigma)

# Plot normal distribution
plt.plot(x, y, label='Normal Distribution (μ=0, σ=1)')
plt.xlabel('X-axis')
plt.ylabel('Probability Density')
plt.title('Normal Distribution')
plt.legend()
plt.grid()
plt.show()
```
